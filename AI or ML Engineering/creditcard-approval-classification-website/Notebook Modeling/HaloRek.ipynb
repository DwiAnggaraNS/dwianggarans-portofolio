Dalam industri perbankan dan keuangan, penilaian kelayakan kredit merupakan salah satu proses yang krusial. ## Pemeriksaan Nilai yang Hilang dan Duplikat

```python
# Menghitung jumlah missing values di setiap kolom pada data_train
def missing_value_table(data):
    missing_values = data.isnull().sum()
    total_rows = len(data)
    percentage_missing = (missing_values / total_rows) * 100
    table = pd.concat([missing_values, percentage_missing], axis=1, keys=['Missing Values', 'Percentage'])
    return table

def calculateRowContainingMissingValues(data):
  total_rows_original_data = len(data)

  # Drop missing values
  data_temp = data.copy()
  data_temp.dropna(inplace=True)
  total_rows_after_deleting_rowMV = len(data_temp)

  # calculate percentage of row that contain missing values
  percentage_row_with_missing_values = (total_rows_original_data - total_rows_after_deleting_rowMV) / total_rows_original_data * 100

  return total_rows_original_data-total_rows_after_deleting_rowMV, total_rows_original_data, percentage_row_with_missing_values

# Call the function
check_missing_value = missing_value_table(df)
print(check_missing_value)
print()

totalrow = calculateRowContainingMissingValues(df)
print(f'Jumlah baris pada data asli = {totalrow[1]}')
print(f'Jumlah baris yang mengandung missing values = {totalrow[0]}')
print(f'Persentase baris yang mengandung missing values terhadap data asli = {round(totalrow[2])}%')
```

Oleh karena itu, kami mempertimbangkan untuk melakukan imputation atau pengisian missing values dengan menggunakan nilai ukuran pemusatan data berdasarkan hasil dari visualisasi distribusi data.


Jika distribusi data fitur numerik menunujukkan berdistribusi normal maka kami melakukan imputation missing values-nya dengan menggunakan nilai mean dari fiturnya. Sedangkan jika distribusi data fitur menunujukkan berdistribusi tidak normal atau terdapat skewness maka kami melakukan imputation missing values-nya dengan menggunakan nilai median dari fiturnya.

* Penggunaan mean untuk imputasi nilai yang hilang pada fitur numerik berdistribusi normal memiliki beberapa keuntungan penting. Mean merupakan estimator tidak bias dan paling efisien secara statistik untuk nilai tengah populasi pada distribusi normal, sehingga dapat mempertahankan karakteristik utama distribusi data (Dong & Peng, 2013). Metode ini konsisten dengan banyak analisis parametrik yang mengasumsikan normalitas, mempertahankan varians asli data, dan relatif mudah diimplementasikan (Enders, 2010). Little & Rubin (2019) juga menegaskan bahwa imputasi dengan mean pada data normal dapat membantu mempertahankan struktur data asli, mendukung asumsi normalitas untuk analisis lanjutan, dan memberikan estimasi yang efisien untuk nilai-nilai yang hilang, yang penting untuk menjaga validitas analisis statistik berikutnya.

* Penggunaan median dalam imputasi nilai yang hilang pada fitur numerik dapat membantu mempertahankan bentuk asli
distribusi data, yang penting untuk
analisis selanjutnya agar hasilnya tidak
bias. Selain itu,penggunaan median dalam kondisi data
yang skewed dapat meningkatkan akurasi
dan validitas analisis data

## Analisis Distribusi Data Variabel Target


```python
palette = ["#ADD8E6","#EFB000"]  # Mendefinisikan palet warna yang akan digunakan dalam visualisasi

# Menghitung nilai persentase dari distribusi persetujuan kartu kredit
l1 = list(df['credit_card_approval'].value_counts())
pie_values = [l1[0] / sum(l1) * 100, l1[1] / sum(l1) * 100]

# Membuat subplots dengan 1 baris dan 2 kolom, serta mengatur ukuran gambar
fig = plt.subplots(nrows = 1,ncols = 2,figsize = (20,7))

# Plot pertama: Diagram Pie
plt.subplot(1,2,1)
plt.pie(pie_values,
        labels = ['Ditolak (0)', 'Disetujui (1)'],  # Label
        autopct = '%1.2f%%',  # Menampilkan persentase pada diagram pie
        startangle = 90,  # Memulai sudut pie dari 90 derajat
        explode = (0.1,0.1),  # Memisahkan kedua bagian pie untuk penekanan
        colors = palette,  # Menerapkan palet warna
        wedgeprops = {'edgecolor': 'black', 'linewidth': 1, 'antialiased': True})  # Mengatur properti wedge
plt.title('Distribusi (%) Persetujuan Kartu Kredit: Ditolak vs Disetujui')  # Judul diagram pie

# Plot kedua: Diagram Batang
plt.subplot(1,2,2)
ax = sns.countplot(data = df,
                   x='credit_card_approval',
                   palette = palette,
                   edgecolor = 'black')  # Menggunakan garis tepi hitam untuk batang
for i in ax.containers:
    ax.bar_label(i,)  # Menambahkan label pada setiap batang
ax.set_xticklabels(['Ditolak (0)', 'Disetujui (1)'])  # Label sumbu X

plt.title('Distribusi Persetujuan Kartu Kredit: Ditolak vs Disetujui')  # Judul diagram batang
plt.show()  # Menampilkan plot
```

Distribusi yang sangat tidak seimbang (imbalanced) ini memberikan beberapa implikasi penting. Distribusi yang sangat tidak seimbang ini dapat menyebabkan bias pada model machine learning, di mana model cenderung memprediksi kelas mayoritas (Rejected) dengan baik tetapi buruk dalam memprediksi kelas minoritas (Approved).

```python
# Calculate outliers for each column
outlier_counts = {}
for column in ['CHILDREN', 'Annual_income', 'Birthday_count', 'Employed_days', 'Family_Members']:
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
  outlier_counts[column] = outliers.shape[0]

# Print outlier counts
print("Outlier Counts:")
for column, count in outlier_counts.items():
  print(f"{column}: {count}")
```

### Uji Hipotesis untuk Fitur Kategorikal
Setiap fitur kategorikal akan diuji secara statistik menggunakan uji hipotesis untuk menentukan apakah fitur tersebut berpengaruh terhadap variabel target atau tidak. Karena variabel target berupa data kategotikal maka uji hipotesis dilakukan dengan menggunakan Chi-Square Test of Independence

1. Pernyataan Hipotesis
  - **Hipotesis Nol (H0):** Tidak ada hubungan antara `fitur kategorikal ke-i` dan `credit approval`. (Distribusi `credit approval` adalah sama di antara semua kategori fitur kategorikal ke-i.)
  - **Hipotesis Alternatif (H1):** Ada hubungan antara `fitur kategorikal ke-i` dan `credit approval`. (Distribusi `credit approval` berbeda di antara kategori `fitur kategorikal ke-i`.)

  Dengan kata lain:

  - **H0:** Persetujuan kredit (credit approval) independen terhadap (fitur kategorikal ke-i).
  - **H1:** Persetujuan kredit (credit approval) tidak independen terhadap (fitur kategorikal ke-i).

2. Tingkat Signifikansi
  - Kami menggunakan tingkat signifikansi α sebesar 5% atau 0.05.

3. Statistik Penguji
  - Kami menggunakan statistik uji Chi-Square, yang dihitung menggunakan pustaka Scipy
4. Daerah Penolakan
  - Untuk menentukan daerah penolakan, kami menggunakan p-value dan   membandingkannya dengan tingkat signifikansi. P-value dihitung menggunakan pustaka Scipy
5. Kesimpulan
Dalam menginterpretasikan hasil Uji Hipotesis, kami menggunakan p-value sebagai dasar pengambilan keputusan:
	- Jika p-value < α (0.05), maka kami menolak hipotesis nol (H0). Ini berarti bahwa fitur kategorikal tersebut memiliki asosiasi yang signifikan dengan variabel target (penerimaan kartu kredit). Dalam konteks penelitian kami, hal ini mengindikasikan bahwa distribusi penerimaan kartu kredit berbeda secara signifikan di antara kategori-kategori dalam fitur tersebut. Dengan kata lain, fitur ini memiliki pengaruh yang bermakna terhadap keputusan penerimaan kartu kredit dan karenanya penting untuk dipertahankan dalam model.
	- Jika p-value ≥ α (0.05), maka kami gagal menolak hipotesis nol (H0). Ini mengindikasikan bahwa tidak ada bukti yang cukup untuk menyatakan adanya hubungan yang signifikan antara fitur kategorikal tersebut dan penerimaan kartu kredit. Dalam hal ini, distribusi penerimaan kartu kredit cenderung serupa di antara kategori-kategori dalam fitur tersebut. Fitur semacam ini mungkin tidak memberikan informasi yang substansial untuk prediksi penerimaan kartu kredit dan dapat dipertimbangkan untuk dihapus dari model demi efisiensi dan simplifikasi.


```python
# Gabungkan kembali X_train dan y_train untuk pengujian chi-square
train_data = X_train.copy()
train_data['credit_card_approval'] = y_train

categorical_columns = [
    'Car_Owner', 'Propert_Owner', 'Type_Income',  'Type_Occupation', 'EDUCATION',
    'Marital_status', 'Housing_type', 'Mobile_phone', 'Work_Phone',
    'Phone', 'EMAIL_ID'
]

def perform_hypothesis_testing_using_chi_square(data_train, categorical_columns):
    alpha = 0.05
    columns_to_remove = []
    for column in categorical_columns:
        contingency_table = pd.crosstab(data_train[column], data_train['credit_card_approval'])
        chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
        print(f"Chi-Square statistic for {column}: {chi2}")
        print(f"p-value for {column}: {p}")
        if p < alpha:
            print(f"Reject the null hypothesis: There is a significant relationship between {column} and 'credit approval'.")
        else:
            print(f"Fail to reject the null hypothesis: There is no significant relationship between {column} and 'credit approval'.")
            columns_to_remove.append(column)
        print("-" * 100)
        print()

    return columns_to_remove

# Panggil fungsi dengan data train
column_to_remove = perform_hypothesis_testing_using_chi_square(train_data, categorical_columns)
print(f'Categorical columns that have no impact on target variable based on hypothesis testing result: {column_to_remove}')

# drop categorical features that have no impact on target variable based on hypothesis testing result dari X_train
for categorical_column in column_to_remove:
    X_train.drop([categorical_column], axis=1, inplace=True)

# Menampilkan hasil
print()
print("Data after Hypothesis Testing for Categorical Features: ")
X_train.head()
```

Berdasarkan hasil tersebut, ditemukan bahwa beberapa fitur kategorik memiliki hubungan signifikan dengan persetujuan kartu kredit, sementara yang lain tidak. Fitur-fitur yang menunjukkan hubungan signifikan (p-value < 0.05) adalah "Type_Income", "Type_Occupation", "EDUCATION", dan "Housing_type". Ini mengindikasikan bahwa fitur-fitur tersebut memiliki pengaruh yang bermakna terhadap keputusan persetujuan kartu kredit.

Fitur-fitur kategorik seperti "Car_Owner", "Propert_Owner", "Marital_status", "Mobile_phone", "Work_Phone", "Phone", dan "EMAIL_ID" gagal menolak hipotesis null (p-value > 0.05). ## Pra-Pemrosesan Data dan Pelatihan Model menggunakan Pipeline

```python
# Identify numeric and categorical columns
numeric_features = ['Annual_income', 'Employed_days', 'Family_Members','Birthday_count']

nominal_features = [
    'Type_Income', 'Housing_type', 'Type_Occupation'
]

ordinal_feature = ['EDUCATION']

# Definisikan urutan pendidikan
education_order = ['Lower secondary', 'Secondary / secondary special', 'Incomplete higher', 'Higher education', 'Academic degree']

# Buat pipeline baru untuk fitur fitur
ordinal_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ordinal', OrdinalEncoder(categories=[education_order], dtype=int))
])

nominal_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(drop='first', sparse=False, handle_unknown='ignore'))
])

numeric_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', RobustScaler()),
])

# Satukan pipeline fitur sebagai data preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_pipeline, numeric_features),
        ('nominal', nominal_pipeline, nominal_features),
        ('ordinal', ordinal_pipeline, ordinal_feature)
    ]
  )

# Define the models with random_state=42
seed = 42
np.random.seed(seed)

# Define models
models = [
    ('model_dt', DecisionTreeClassifier(random_state=seed)),
    ('model_knn', KNeighborsClassifier()),
    ('model_gnb', GaussianNB()),
    ('model_rf', RandomForestClassifier(random_state=seed)),
    ('model_xgb', XGBClassifier(random_state=seed)),
    ('model_et', ExtraTreesClassifier(random_state=seed)),
    ('model_gbc', GradientBoostingClassifier(random_state=seed)),
]

# Dictionary to store the results and trained models
results_df = pd.DataFrame(columns=["Model","Test AUC"])
trained_models = {}

# Fungsi untuk mengevaluasi model
def evaluate_model(pipeline, X, y):
    y_pred = pipeline.predict(X)

    if hasattr(pipeline, "predict_proba"):
        y_pred_proba = pipeline.predict_proba(X)[:, 1]
        auc = roc_auc_score(y, y_pred_proba)
    else:
        auc = 'N/A'

    return auc

for name, model in models:
    # Create complete pipeline with SMOTE
    pipeline = ImbPipeline([
        ('preprocessor', preprocessor),
        ('classifier', model)
    ])

    # Train the model on the training set
    pipeline.fit(X_train, y_train)
    trained_models[name] = pipeline

    # Evaluate on the test set
    test_auc = evaluate_model(pipeline, X_test, y_test)

    # Append the results to the DataFrame
    results_df.loc[len(results_df)] = [name, test_auc]

# Display the results
print(results_df)
```

Hasil pelatihan dan evaluasi di atas menunjukkan bahwa model Extra Trees Classifier memiliki performa terbaik dengan nilai AUC tertinggi sebesar 0,887798, diikuti oleh Random Forest Classifier dengan AUC 0,856726 dan XGBoost dengan AUC 0,810893. Model Extra Trees Classifier mengungguli model lainnya dalam hal kemampuan diskriminatif dan generalisasi.
```python
# Now separate features and target variables in each dataset
X_train = df_train.drop(['label', 'id', 'attack_cat'], axis=1)
y_train = df_train['label']

X_test = df_test.drop(['label', 'id', 'attack_cat'], axis=1)
y_test = df_test['label']

# Define preprocessing pipelines
nominal_pipeline = Pipeline([
    ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
])

numeric_pipeline = Pipeline([
    ('scaler', MinMaxScaler()),
])

# Combine feature pipelines for data preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_pipeline, selected_numeric_features),
        ('nominal', nominal_pipeline, selected_categorical_features),
    ],
    remainder='passthrough'  # Handle any columns not explicitly selected
)

# Set seed for reproducibility
seed = 42
np.random.seed(seed)

# Define classification models to evaluate
models = [
    ('model_lr', LogisticRegression(random_state=seed, max_iter=1000)),
    ('model_dt', DecisionTreeClassifier(random_state=seed)),
    ('model_knn', KNeighborsClassifier()),
    ('model_rf', RandomForestClassifier(random_state=seed, n_jobs=-1)),
    ('model_xgb', XGBClassifier(random_state=seed, use_label_encoder=False, eval_metric='logloss')),
    ('model_lgbm', LGBMClassifier(random_state=seed, verbose=-1)),
    ('model_et', ExtraTreesClassifier(random_state=seed, n_jobs=-1)),
    ('model_gbc', GradientBoostingClassifier(random_state=seed)),
    ('model_ada', AdaBoostClassifier(random_state=seed)),
]

# Initialize results DataFrame with columns for both train and test metrics
results_df = pd.DataFrame(columns=[
    "Model",
    "Train AUC", "Train Accuracy", "Train Precision", "Train Recall", "Train F1-score",
    "Test AUC", "Test Accuracy", "Test Precision", "Test Recall", "Test F1-score"
])

trained_models = {}

# Function to evaluate model performance
def evaluate_model(pipeline, X, y_true):
    """Evaluate model performance on given data"""
    y_pred = pipeline.predict(X)

    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, zero_division=0)
    recall = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)

    # Calculate AUC if model supports probability predictions
    auc = 'N/A'
    if hasattr(pipeline, "predict_proba"):
        try:
            y_pred_proba = pipeline.predict_proba(X)[:, 1]
            auc = roc_auc_score(y_true, y_pred_proba)
        except Exception as e:
            print(f"Could not calculate AUC: {e}")

    return auc, accuracy, precision, recall, f1

# Training and evaluation loop
print("Training and evaluating models...")
for name, model in models:
    print(f"Processing {name}...")

    try:
        # Create pipeline with preprocessing and classifier
        pipeline = Pipeline([
            ('preprocessor', preprocessor),
            ('classifier', model)
        ])

        # Train the model
        pipeline.fit(X_train, y_train)
        trained_models[name] = pipeline

        # Evaluate on training data
        train_metrics = evaluate_model(pipeline, X_train, y_train)

        # Evaluate on test data
        test_metrics = evaluate_model(pipeline, X_test, y_test)

        # Combine results
        model_results = [name] + list(train_metrics) + list(test_metrics)
        results_df.loc[len(results_df)] = model_results

    except Exception as e:
        print(f"Error processing {name}: {str(e)}")
        # Add error entry to results
        results_df.loc[len(results_df)] = [name] + ['Error']*10

# Display the results with clear separation between train and test metrics
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
print("\n--- Model Evaluation Results ---")
print(results_df)

# Convert metrics to numeric for comparison
for col in results_df.columns[1:]:
    results_df[col] = pd.to_numeric(results_df[col], errors='coerce')

# Find best model based on test F1-score
if not results_df.empty and 'Test F1-score' in results_df.columns and results_df['Test F1-score'].notna().any():
    best_idx = results_df['Test F1-score'].idxmax()
    best_model_name = results_df.loc[best_idx, 'Model']
    best_model_pipeline = trained_models.get(best_model_name)

    print(f"\nBest model (by Test F1-score): {best_model_name}")
    print(f"Best model metrics:")
    for metric in ['AUC', 'Accuracy', 'Precision', 'Recall', 'F1-score']:
        train_val = results_df.loc[best_idx, f'Train {metric}']
        test_val = results_df.loc[best_idx, f'Test {metric}']
        diff = train_val - test_val
        print(f"  {metric}: Train={train_val:.4f}, Test={test_val:.4f}, Diff={diff:.4f}")

    # Overfitting analysis
    avg_diff = np.mean([
        results_df.loc[best_idx, f'Train {m}'] - results_df.loc[best_idx, f'Test {m}']
        for m in ['AUC', 'Accuracy', 'Precision', 'Recall', 'F1-score']
        if not np.isnan(results_df.loc[best_idx, f'Train {m}']) and not np.isnan(results_df.loc[best_idx, f'Test {m}'])
    ])

    print("\nModel fit analysis:")
    if avg_diff > 0.1:
        print("  Warning: Potential overfitting (train metrics significantly better than test metrics)")
    elif avg_diff < -0.05:
        print("  Warning: Unusual pattern - test metrics better than train metrics")
    else:
        print("  Good fit: Train and test metrics are similar")

else:
    print("\nCould not determine best model.")

# Optional: Visualize train vs test metrics for all models
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot F1-scores
ax = axes[0]
model_names = results_df['Model'].str.replace('model_', '')
train_f1 = results_df['Train F1-score']
test_f1 = results_df['Test F1-score']

x = np.arange(len(model_names))
width = 0.35

ax.bar(x - width/2, train_f1, width, label='Train F1-Score')
ax.bar(x + width/2, test_f1, width, label='Test F1-Score')
ax.set_title('Train vs Test F1-Score by Model')
ax.set_xticks(x)
ax.set_xticklabels(model_names, rotation=45, ha='right')
ax.set_ylabel('F1-Score')
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Plot difference between train and test (to show overfitting)
ax = axes[1]
f1_diff = train_f1 - test_f1
colors = ['red' if x > 0.1 else 'green' if x < 0.05 else 'blue' for x in f1_diff]

ax.bar(x, f1_diff, color=colors)
ax.set_title('Train/Test F1-Score Difference (Potential Overfitting)')
ax.set_xticks(x)
ax.set_xticklabels(model_names, rotation=45, ha='right')
ax.set_ylabel('Train F1 - Test F1')
ax.axhline(y=0.1, color='r', linestyle='--', alpha=0.7, label='Potential Overfitting Threshold')
ax.axhline(y=0, color='g', linestyle='-', alpha=0.7)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
